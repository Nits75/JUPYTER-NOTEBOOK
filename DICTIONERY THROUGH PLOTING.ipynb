```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
dict={"name":["sachin","virat","pandya","sehvag","dhoni","rohit"],
     "year":[2020,2021,2018,2017,2015,2010],
     "run":[200,100,50,60,80,150]
      }
dict=pd.DataFrame(dict)
dict
```

```python
style.use('ggplot')
plt.title("indian cricet team")
plt.xlabel("players")
plt.ylabel("runs")
plt.plot(dict.run,'.')
plt.show()
```

```python
plt.bar(dict.name,dict.run)
plt.show()
```

```python
dict[0:100].plot('name','run')
plt.show()
```

```python
dict[0:10].plot.box()
plt.show()
```

```python
dict[0:10].plot(kind='scatter', x='name',y='run')
plt.show()
```

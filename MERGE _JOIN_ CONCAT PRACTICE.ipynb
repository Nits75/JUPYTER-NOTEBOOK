```python
import pandas as pd 
import numpy as np 
```

```python
city1 = {"city":["mumbai","banglore","delhi"],
             "years":[189,223,111]}
city2 = {"city":["mumbai","banglore","delhi"],
              "years":[200,889,789]}
```

```python
df1=pd.DataFrame(city1)
df1
```

```python
df2=pd.DataFrame(city2)
df2
```

```python
merge_ = pd.merge(df1,df2,on="city")
```

```python
merge_
```

```python
city3={"city":["panjim1","vasco2","madgaon3"],
        "years":[200,457,877]}
city4={"city":["panjim","vasco","madgaon"],
         "years":[500,600,550]}

df3 = pd.DataFrame(city3)
df4 = pd.DataFrame(city4)
```

```python
merge_=pd.merge(df3,df4,how="left",on="city")  #left join
merge_
```

```python
merge_=pd.merge(df3,df4,how="right",on="city") #right join 
merge_
```

```python
merge_=pd.merge(df3,df4,how="outer",on="city") #full merge 
merge_
```

```python
merge_=pd.merge(df3,df4,how="outer",on="city",indicator="true")
merge_
```

#join

```python
left_df=pd.DataFrame({'A':['A4','A7','A2'],
                       'B':['A4','A7','A2']},
                     index=pd.Index(['0', '1', '2']))
                    
```

```python
right_df=pd.DataFrame({'C':['C0','C1','C2'],
                        'D':['D1','D2','D3']},
                     index=pd.Index(['0','1','2']))
```

```python
result= left_df.join(right_df, how='inner')
result
```

```python
result = left_df.join(right_df, how='outer')
result
```
